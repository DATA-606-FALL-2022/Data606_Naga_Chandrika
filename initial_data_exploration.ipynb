## Top 15 crimes that had happened in LOs Angeles

```python
CrimeAnalysis_df['Crm Cd Desc'].value_counts().head(15).plot.bar( figsize = (20, 10))
plt.title('Top 15 Crimes in Los Angeles', fontweight = 30, fontsize = 20)
plt.show()
```

## Finding the particular area where crimes are high

```python
CrimeAnalysis_df['AREA NAME'].value_counts().plot.bar(figsize = (20, 10))
plt.title('Total Crimes in particular area',fontsize = 30)
plt.show()
```
